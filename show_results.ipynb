```python
# Son 17 DataSets, 9 PreModels repartidos en 50 combinaciones/conjuntos de 3 a 7 elementos + conjunto vacio
# de la columna 0..6 corresponde al correr: train + predict + test 
# de la columna 7 en adelante es la validación cruzada solo con el train
# combinations: arreglo con los PreModels de EvoMSA usados en el training
# combina_1 y combina_2 son los índices del arreglo de PreModels. Es con lo que se enlaza el test y la validacíon cruzada
# en cada renglón, se excluye el PreModel que corresponde al Dataset que esta siendo probado

df_.head()
```

```python
ds_final = pd.DataFrame()
df_.rename(columns = {'mean':'media', 'max':'maximo', 'min':'minimo'}, inplace = True)
```

```python
for ds_n in df_.dataset.unique():
    _ = df_.loc[df_['dataset'] == ds_n].sort_values(by=['media'], ascending=False)
    empty_value = _.loc[_['combina_1'] == 50].media.iat[0]
    dse = _.loc[(_['media'] > empty_value) | (_['combina_1'] == 50)]    
    dse['diff'] = dse.apply(lambda row: row.media - empty_value, axis=1) 
    dse = dse.loc[dse['diff'] > 0.01]  
    ds_final = pd.concat([ds_final, dse], ignore_index=True)        
ds_final     
```

```python
_ = df_.loc[df_['dataset'] == 'semeval2018_anger'].sort_values(by=['f1'], ascending=False)
empty_f1 = _.loc[_['combina_1'] == 0].f1.iat[0]
dse = _.loc[(_['f1'] > empty_f1) | (_['combina_1'] == 0)]    
dse['diff_f1'] = dse.apply(lambda row: row.f1 - empty_f1, axis=1) 
dse = dse.loc[(dse['diff_f1'] > 0.01) | (_['combina_1'] == 0)] 
print(empty_f1, len(dse))
dse.style.background_gradient(cmap="coolwarm", subset=['f1', 'media', 'diff_f1'])
```
